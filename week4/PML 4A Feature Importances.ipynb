````python
import numpy as np
from sklearn import datasets
from sklearn import tree
tree_model = tree.DecisionTreeClassifier()
iris = datasets.load_iris()
````

````python
np.random.seed(2)
tree_model.set_params(max_depth=2)
tree_model.fit(iris.data[:, 2:4], iris.target)
tree_model.score(iris.data[:, 2:4], iris.target)
````

Visualizing decision trees typically requires at least one additional package to be installed.
If you are using Anaconda, then you should install:

```shell
conda install python-graphviz
```

````python
import graphviz
dot_data = tree.export_graphviz(tree_model, out_file=None)
graph = graphviz.Source(dot_data)
graph.render('pic/iris.png')
graph
````

We can generate more appealing decision trees by using export_graphviz options:

````python
import graphviz
dot_data = tree.export_graphviz(
    tree_model,
    filled=True,
    rounded=True,
    class_names=iris.target_names,
    feature_names=iris.feature_names[2:4],
    out_file=None
)
graph = graphviz.Source(dot_data)
graph.render('pic/iris_enhanced.png')
graph
````
